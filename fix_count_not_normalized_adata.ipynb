```python
import anndata as ad
import scanpy as sc
```

```python
raw_adata =  ad.read_h5ad('DATA/GSE155249_RAW.h5ad')
raw_adata
```

```python
main_adata = ad.read_h5ad('DATA/GSE155249_main.h5ad')
main_adata
```

```python
raw_adata.obs['Sample']
```

```python
common_obs = list(set(raw_adata.obs_names).intersection(main_adata.obs_names))
len(common_obs)
```

```python
adata_raw_human = [x for x in raw_adata.var_names if x.startswith('GRCh38')]
print('liczba ludzkich genow '+ str(len(adata_raw_human)))
adata_raw_human = [x for x in adata_raw_human if not x.startswith('GRCh38______AL')]
print('liczba ludzkich genow bez tych zaczynajacych sie od GRCh38______AL = ' + str(len(adata_raw_human)))
```

```python
adata_raw_human_id = [x[12:] for x in adata_raw_human]
adata_raw_human_id

common_genes = list(set(main_adata.var_names).intersection(adata_raw_human_id))
print('Liczba genow w naszym MAIN = ' + str(len(main_adata.var)))
print('Liczba wspolnych genow pomiedzy RAW i MAIN = ' + str(len(common_genes)))
```

```python
common_genes = ['GRCh38______' + x for x in common_genes]
df = sc.get.obs_df(adata=raw_adata, keys=common_genes)
df
```

```python
df_final = df[[x in common_obs for x in df.index]]
df_final.head(5)
```

```python
all((df_final.index == main_adata.obs.index) == True)
```

```python
adata_with_counts = sc.AnnData(df_final, df_final.index.to_frame(), df_final.columns.to_frame())
adata_with_counts.obs.rename(columns={0:"barcode_name"}, inplace=True)
adata_with_counts.var.rename(columns={0:"gene_name"}, inplace=True)
adata_with_counts
```

```python
all((adata_with_counts.obs.index == main_adata.obs.index) == True)
```

```python
adata_with_counts.obs['Sample'] = main_adata.obs['Sample']
adata_with_counts.obs['Cluster'] = main_adata.obs['Cluster']
adata_with_counts.layers["counts"] = adata_with_counts.X.copy()
adata_with_counts
```

```python
sc.write('DATA/GSE155249_COUNTS_NOT_NORMALIZED_FIXED.h5ad', adata_with_counts)
```
