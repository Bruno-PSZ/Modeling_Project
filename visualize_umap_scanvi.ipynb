```python
import anndata as ad
import scanpy as sc
```

```python
adata =  ad.read_h5ad('DATA/ANNDATAS/adata_scanvi_4369.h5ad')
adata
```

```python
SCANVI_LATENT_KEY = "X_scANVI"
```

```python
sc.pp.neighbors(adata, use_rep=SCANVI_LATENT_KEY)
sc.tl.umap(adata, min_dist=0.3)
```

```python
fig = sc.pl.umap(
    adata,
    color=["Cluster"],
    frameon=False,
    ncols=1,
    return_fig=True,
    legend_loc="on data",       
    legend_fontsize=5       
)

fig.suptitle("UMAP of SCANVI Representation", fontsize=16)
```
